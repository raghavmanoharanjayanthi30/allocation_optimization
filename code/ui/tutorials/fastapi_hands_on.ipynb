# FastAPI Hands-On\n\n1. Start backend from project root:\n` .venv/bin/python -m uvicorn code.ui.backend.app.main:app --reload --port 8000 `\n2. Run the next cell to call your API.

```python
import requests\n\nbase = "http://localhost:8000"\nprint("health:", requests.get(f"{base}/health", timeout=10).json())\nprint("methods:", requests.get(f"{base}/methods", timeout=10).json())
```

```python
payload = {\n    "technician_count": 8,\n    "job_count": 6,\n    "technician_seed": 101,\n    "job_seed": 202\n}\nscenario = requests.post(f"{base}/generate-scenario", json=payload, timeout=30).json()\nprint("technicians:", len(scenario["technicians"]))\nprint("jobs:", len(scenario["jobs"]))
```

```python
alloc = {\n    "method": "greedy",\n    "technicians": scenario["technicians"],\n    "jobs": scenario["jobs"],\n}\nresult = requests.post(f"{base}/allocate", json=alloc, timeout=30).json()\nresult
```
